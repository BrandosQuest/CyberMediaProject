```python
import demucs.api
import soundfile as sf
import torch
import musdb
import IPython.display as ipd
import os
```

```python
noisy = False
mus = musdb.DB(download=True)
track = mus[0]

choice = input("Do you want to load a noisy track? (y/n): ")

if choice.lower() == "y":
    noisy_path = './songs/testSample_noisy.wav'
    if os.path.exists(noisy_path):
        audio, sr = sf.read(noisy_path)
        noisy = True
    else:
        print("File not found! Loading clean track instead...")
        audio = track.audio
        sr = track.rate
elif choice.lower() == "n":
    audio = track.audio
    sr = track.rate
else:
    audio = track.audio
    sr = track.rate

# Convert to PyTorch tensor for Demucs ([samples, channels] -> [channels, samples])
wav_tensor = torch.tensor(audio).T.float()

# Demucs requires stereo audio. If mono, duplicate the channel to fake stereo.
if wav_tensor.ndim == 1:
    wav_tensor = wav_tensor.unsqueeze(0).repeat(2, 1)

print(f"Audio loaded! Tensor shape: {wav_tensor.shape} | Sample Rate: {sr} Hz")
```
