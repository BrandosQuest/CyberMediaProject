```python
import demucs.api
import soundfile as sf
import torch
import IPython.display as ipd
```

```python
print("1. Loading Demucs model...")
separator = demucs.api.Separator()

print("2. Reading file...")
audio_data, sample_rate = sf.read("./songs/test.mp3")

wav_tensor = torch.tensor(audio_data).T.float()

if wav_tensor.ndim == 1:
    wav_tensor = wav_tensor.unsqueeze(0).repeat(2, 1)

print("3. Please wait, this could take a minute...")
origin, separated = separator.separate_tensor(wav_tensor, sample_rate)
```

```python
for name, tensor in separated.items():
    audio_numpy = tensor.numpy()

    print(f"Displaying {name}")
    ipd.display(ipd.Audio(audio_numpy, rate=sample_rate))
    
    name_file = f"{name}_separato.wav"
    sf.write("./output/" + name_file, audio_numpy.T, sample_rate, format='WAV')
```
