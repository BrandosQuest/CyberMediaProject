```python
import numpy as np
import librosa
import IPython.display as ipd
import matplotlib.pyplot as plt

# 1. Caricamento del file audio
audio_path = "songs/test.mp3"  # Sostituisci con il tuo file
y, sr = librosa.load(audio_path, sr=None, mono=True)

print(f"Audio caricato con successo. Frequenza di campionamento: {sr} Hz")
```

```python
# Calcolo della STFT complessa
X = librosa.stft(y)
print("Calcolo della STFT complessa avviato")

# Decomposizione in componenti armoniche (X_h) e percussive (X_p)
# Passando direttamente la matrice complessa X, librosa restituisce maschere complesse
X_h, X_p = librosa.decompose.hpss(X, margin=1.2)

print("Decomposizione HPSS completata!")
```

```python
# iSTFT per ottenere le forme d'onda
y_harmonic = librosa.istft(X_h, length=len(y))
y_percussive = librosa.istft(X_p, length=len(y))

# Playback
print("Traccia Originale:")
display(ipd.Audio(y, rate=sr))

print("Traccia Armonica (Vocals, Synth, Archi...):")
display(ipd.Audio(y_harmonic, rate=sr))

print("Traccia Percussiva (Batteria, Kick, Snare...):")
display(ipd.Audio(y_percussive, rate=sr))

# RICORDA: Fai "Edit > Clear All Outputs" prima di fare il commit su GitHub!

#y_harmonic, y_percussive = librosa.effects.hpss(y)
```
